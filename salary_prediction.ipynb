```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df['experience'] = df['experience'].fillna('zero')
df['experience']
```

```python
word_to_number = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
}

df['experience'] = df['experience'].map(word_to_number)
df['experience']
```

```python
df
```

```python
score_median = df['test_score(out of 10)'].median()
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(score_median)
```

```python
df['test_score(out of 10)']
```

```python
df
```

```python
new_df = df.drop('salary($)',axis='columns')
new_df
```

```python
salary = df['salary($)']
salary
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,salary)
```
